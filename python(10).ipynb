```python
import random
import numpy as np
grid_size=2
start_point=(0,0)
goal=(2,2)
actions=["left","right","up","down"]
def move(state,action):
   x,y=state
   if action==0:
     y=y-1
   elif action==1:
     y=y+1
   elif action==2:
     x=x-1
   else:
     x=x+1
   x=max(0,min(grid_size,x))
   y=max(0,min(grid_size,y))

   next_state=x,y

   if next_state==goal:
      return next_state,10,True
   elif next_state==pit:
      return next_state,-10,True
   else:
      return next_state,-1,False
Q=np.zeros((grid_size+1,grid_size+1,len(actions)))
alpha=0.5
gamma=0.9
epsilon=0.2
episode=500
for i in range(episode):
    state=start_point
    done=False
    while not done:
     x,y=state
     if random.random()<epsilon:
      action=random.randint(0,3)
     else:
      action=np.argmax(Q[x,y])
    next_state,reward,done=move(state,action)
    x1,y1=next_state
    Q[x,y,action]=Q[x,y,action]+alpha*(reward+gamma*np.max(Q[x1,y1])-Q[x,y,action])
    state=next_state
print("here is learned policy")
for i in range(3):
  row=[]
  for j in range(3):
    if (i,j)==goal:
       row.append("G")
    elif (i,j)==pit:
       row.append("*")
    else:
       best_action=actions[np.argmax(Q[i,j])]
       row.append(best_action[0])
  print(row)
```

```python
from abc import ABC,abstractmethod

class Payment(ABC):
  @abstractmethod
  def payment(self,method,amount):
    pass
class UpiPayment(Payment):
      def payment(self,method,amount):
        print(f"{amount}done by{method}done")
class CreditCardPayment(Payment):
      def payment(self,method,amount):
        print(f"{amount}done by{method}done")
class DebitCardPayment(Payment):
      def payment(self,method,amount):
           print(f"{amount}done by{method}done")
object1=UpiPayment()
object2=CreditCardPayment()
object3=DebitCardPayment()

def payment_process(method):
  if method=="Upi":
    return UpiPayment()
  elif method=="Credit":
    return CreditCardPayment()
  else:
    return DebitCardPayment()
pay_object=payment_process("Upi")
pay_object.payment("Gpay",10000)
```
