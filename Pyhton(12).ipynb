```python
#Basic mini calculater for two numbers with basic operators
num1=int(input("enter num1:"))
num2=int(input("enter num2:"))
choices=input("enter the operator(+,-,*,/,**,%)")
if "+" in choices:
  print(num1+num2) #add numbers
elif "-" in choices:
  print(num1-num2) #subtract numbers
elif "*" in choices:
  print(num1*num2) #multiply numbers
elif "/" in choices:
  print(num1/num2) #divide numbers
elif "**" in choices:
  print(num1**num2) #find the power value
elif "%" in choices:
  print(num1%num2) #find remender after dividing the numbers
else:
  print("invalid operator")
```
