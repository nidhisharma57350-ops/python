```python
customers_name=["Mohit","Rajesh","Manish","Komal","Nidhi","Manish","Gourav","kritika","Lalit","Sunidhi"]
states=["Hp","Punjab","Hariyana","Hp","Punjab","Mp","Hp","Up","Bihar","Hariyana"]
products=["mobile","ac","mobile","led","mobile","led","ac","refrigrator","mobile","led"]
genders=["male","male","male","female","female","male","male","female","male","female"]
amount=[10000,40000,20000,30000,40000,50000,60000,20000,15000,30000]
costs=[8000,36000,15000,28000,36000,45000,55000,18000,12000,25000]
unique_states=[]
state_wise_sale=[]
for i in states:
  if i not in unique_states:
    unique_states.append(i)
    state_wise_sale.append(0)
for i in range(len(states)):
  index=unique_states.index(states[i])
  state_wise_sale[index]=state_wise_sale[index]+amount[i]
print(unique_states)
print(state_wise_sale)
import matplotlib.pyplot as plt
plt.bar(unique_states,state_wise_sale)
plt.show()
unique_products=[]
product_wise_sale=[]
for i in products:
  if i not in unique_products:
    unique_products.append(i)
    product_wise_sale.append(0)
for i in range(len(products)):
  index=unique_products.index(products[i])
  product_wise_sale[index]=product_wise_sale[index]+amount[i]
print(unique_products)
print(product_wise_sale)
import matplotlib.pyplot as plt
plt.bar(unique_products,product_wise_sale)
plt.show()
unique_genders=[]
genders_wise_sale=[]
for i in genders:
  if i not in unique_genders:
    unique_genders.append(i)
    genders_wise_sale.append(0)
for i in range(len(genders)):
    index=unique_genders.index(genders[i])
    genders_wise_sale[index]=genders_wise_sale[index]+amount[i]
print(unique_genders)
print(genders_wise_sale)
import matplotlib.pyplot as plt
plt.bar(unique_genders,genders_wise_sale)
plt.show()
unique_states=[]
profits=[]
states_wise_profit=[]
for i in states:
  if i not in unique_states:
    unique_states.append(i)
    states_wise_profit.append(0)
for i in range(len(states)):
    profit=amount[i]-costs[i]
    profits.append(profit)
    index=unique_states.index(states[i])
    states_wise_profit[index]=states_wise_profit[index]+profits[i]
print(unique_states)
print(states_wise_profit)
import matplotlib.pyplot as plt
plt.bar(unique_states,states_wise_profit)
plt.show()
```
