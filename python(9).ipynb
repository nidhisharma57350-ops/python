```python
import csv
with open("sales.csv","r")as file:
    reader=csv.reader(file)
    for row in reader:
      print(row)
import pandas as pd
df=pd.read_csv("sales.csv")
df["final_price"] = df["Amount"]*df["quantity"]*(1 - df["discount"]/100)
print(df)
df.to_csv("final_price.csv")
```

```python
import csv
with open("sales_data.csv","r")as file:
    reader=csv.reader(file)
    for row in reader:
      print(row)
import pandas as pd
df=pd.read_csv("sales_data.csv")
df["final_price"] = df["Amount"]*df["quantity"]*(1 - df["discount"]/100)
dx=df[df["final_price"]>50000]
print(dx)
dx.to_csv(">50000.csv")
```
