```python
import csv
name,gender,location,product,amount=[],[],[],[],[]
with open("sales.csv","r")as file:
   reader=csv.reader(file)
   header=next(reader)
   for row in reader:
     #print(row)
     name.append(row[0])
     gender.append(row[1])
     location.append(row[2])
     product.append(row[3])
     amount.append(int(row[4]))
print(name)
print(gender)
print(location)
print(product)
print(amount)
unique_location=[]
location_wise_sale=[]
for i in location:
   if i  not in unique_location:
     unique_location.append(i)
     location_wise_sale.append(0)
for i in range(len(location)):
  index1=unique_location.index(location[i])
  location_wise_sale[index1]=location_wise_sale[index1]+amount[i]
print(unique_location)
print(location_wise_sale)
import matplotlib.pyplot as plt
plt.bar(unique_location,location_wise_sale)
plt.show()
unique_products=[]
product_wise_sale=[]
for i in product:
   if i not in unique_products:
     unique_products.append(i)
     product_wise_sale.append(0)
for i in range(len(product)):
   index2=unique_products.index(product[i])
   product_wise_sale[index2]=product_wise_sale[index2]+amount[i]
print(unique_products)
print(product_wise_sale)
import matplotlib.pyplot as plt
plt.bar(unique_products,product_wise_sale)
plt.show()
unique_genders=[]
gender_wise_sale=[]
for i in gender:
   if i not in unique_genders:
    unique_genders.append(i)
    gender_wise_sale.append(0)
for i in range(len(gender)):
    index3=unique_genders.index(gender[i])
    gender_wise_sale[index3]=gender_wise_sale[index3]+amount[i]
print(unique_genders)
print(gender_wise_sale)
import matplotlib.pyplot as plt
plt.bar(unique_genders,gender_wise_sale)
plt.show()
```
